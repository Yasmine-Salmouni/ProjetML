```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Création d'un DataFrame fictif
df = pd.DataFrame({
    'x': range(10),
    'y': [i**2 for i in range(10)]
})
df.head()
```

```python
# Visualisation simple
plt.plot(df['x'], df['y'])
plt.title('Courbe quadratique')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
